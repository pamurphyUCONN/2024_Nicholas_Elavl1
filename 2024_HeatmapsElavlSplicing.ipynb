```python
#This will filter the metadatafile for specific column names
import pandas as pd

# Read the TSV file into a DataFrame
df = pd.read_csv('/Users/patrickmurphy/Desktop/LeafcutterHeatmapElavl/LigVContraRibo/out/leafcutter_results.tsv', delimiter='\t', skiprows=1, header=0)

# Define the original list of column names
original_column_names = ["Elavl1_ECKO_23_noPCSK9", "Elavl1_ECKO_26", "Elavl1_ECKO_Contra_Rep3", "Elavl1_ECKO_Contra_Rep5", "Elavl1_ECKO_Lig_Rep2", "Elavl1_ECKO_Lig_Rep3", "Elavl1_ECKO_Lig_Rep4", "Elavl1_ECKO_Lig_Rep5", "Elavl1_ECWT_21_noPCSK9", "Elavl1_ECWT_24", "Elavl1_ECWT_Contra_Rep4", "Elavl1_ECWT_Contra_Rep5", "Elavl1_ECWT_Lig_Rep2", "Elavl1_ECWT_Lig_Rep3", "Elavl1_ECWT_Lig_Rep4", "Elavl1_ECWT_Lig_Rep5", "L-bound-11500", "L-bound-11517", "L-bound-11519", "L-bound-11523", "L-bound-11524", "L-bound-11535", "L-bound-11537", "L-bound-11538", "L-bound-11539", "L-bound-11540", "L-bound-11541", "L-unbound-11500", "L-unbound-11519", "L-unbound-11537", "R-bound-11500", "R-bound-11517", "R-bound-11519", "R-bound-11523", "R-bound-11524", "R-bound-11535", "R-bound-11537", "R-bound-11538", "R-bound-11539", "R-bound-11540", "R-bound-11541", "R-unbound-11500", "R-unbound-11519", "R-unbound-11537", "mAECS_12205noTNF", "mAECS_12206noTNF", "mAECS_12258noTNF", "mAECS_12261noTNF", "mAECS_12310noTNF", "mAECS_12311noTNF", "mAECS_12315noTNF", "mAECS_12316noTNF", "mAECS_12368noTNF", "mAECS_12369noTNF", "mAECS_12371noTNF", "mAECS_12374noTNF", "mAECs_12205plusTNF", "mAECs_12206plusTNF", "mAECs_12258plusTNF", "mAECs_12261plusTNF", "mAECs_12310plusTNF", "mAECs_12311plusTNF", "mAECs_12315plusTNF", "mAECs_12316plusTNF", "mAECs_12368plusTNF", "mAECs_12369plusTNF", "mAECs_12371plusTNF", "mAECs_12374plusTNF"]

# Define the names you want to keep (from the metadata files that will be used, which is from the comp file)
names_to_keep = ["L-bound-11535", "L-bound-11537", "L-bound-11538", "L-bound-11539", "L-bound-11540", "L-bound-11541", "R-bound-11535", "R-bound-11537", "R-bound-11538", "R-bound-11539", "R-bound-11540", "R-bound-11541"]

# Find the indices of the names you want to keep in the original list
indices_to_keep = [original_column_names.index(name) for name in names_to_keep]

# Define the function to filter numbers based on the desired indices
def filter_numbers(row, indices):
    values = row.split(',')
    new_values = [values[i] for i in indices]
    return ','.join(new_values)

# Apply the function to the 'PSI' column
df['PSI'] = df['PSI'].apply(lambda x: filter_numbers(x, indices_to_keep))

# Apply the function to the 'ReadCount1' column
df['ReadCount1'] = df['ReadCount1'].apply(lambda x: filter_numbers(x, indices_to_keep))

# Export the DataFrame to a new TSV file
df.to_csv('/Users/patrickmurphy/Desktop/LeafcutterHeatmapElavl/LigVContraRibo/out/leafcutter_results.tsv', sep='\t', index=False)
```
